```python
from qiskit import QuantumCircuit , transpile 
from qiskit_aer import AerSimulator 
import pennylane as qml 
from pennylane import numpy as np 

# device setup
n_qubits = 2 
dev = qml.device("default.qubit" , wires = n_qubits) # means to abits and internal convoluted wire is actually 2 

## quantumn autoencoder model
@qml.qnode(dev)
## defining fucntions 

def qae(inputs):

    qml.RY(inputs[0] , wires = 0 )## rotattion aling y axis using oauli y gate 
    qml.CNOT(wires= [0,1])
    qml.RZ(inputs[1], wires = 1 )
    return qml.probs(wires=[0 ,1])


## Input states and run 
inputs = ([np.pi/4 , np.pi/2])
probs = qae(inputs)
print("Auto encoder probabilities:" , probs)



```
