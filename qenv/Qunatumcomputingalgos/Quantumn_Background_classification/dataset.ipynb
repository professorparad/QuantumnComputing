```python
# so no missing data 
# performing Exploitory data analysis for different values 
import math 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np 
a= []
b=[]
f_length = np.array(df.fLength)
f_width = np.array(df.fWidth)
fSize = np.array(df.fSize)
fConc =  np.array(df.fConc)
fConc1=np.array(df.fConc1)
fAsym =np.array(df.fAsym)
fM3Long=np.array(df.fM3Long)
fM3Trans=np.array(df.fM3Trans)
fAlpha=np.array(df.fAlpha)
fDist =np.array(df.fDist)
y_label = np.array(df.label)
print(y_label)
for i in y_label :
    if i == 1 :
        a.append(i)
    else :
        b.append(i)
y_signal = np.array(a)
y_background = np.array(b)
print(y_signal.size)
print(y_background)


# ---------- KDE for all features in one large frame ----------
features = [c for c in df.columns if c != "label"]
n = len(features)
ncols = 3
nrows = math.ceil(n / ncols)

plt.style.use("seaborn-v0_8-whitegrid")
fig, axes = plt.subplots(nrows, ncols, figsize=(18, 4.8 * nrows))
axes = axes.flatten()

for i, feat in enumerate(features):
    ax = axes[i]
    sns.kdeplot(
        data=df, x=feat, hue="label",
        fill=True, common_norm=False, alpha=0.30, linewidth=1.5, ax=ax
    )
    ax.set_title(f"KDE: {feat}")
    ax.set_xlabel(feat)
    ax.set_ylabel("Density")
    ax.legend(title="Class")

# Hide empty axes (if any)
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

fig.suptitle("KDE Distributions: Signal vs Background (All Features).png", fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
fig.savefig(
    "/home/adminpc/Documents/Professors_Diary/QuantumnComputing/qenv/Qunatumcomputingalgos/Quantumn_Background_classification/plots",
    dpi=300, bbox_inches="tight"
)

```
