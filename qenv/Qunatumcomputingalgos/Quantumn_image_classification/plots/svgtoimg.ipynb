```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plot 
```

```python
freq = np.array([742.06 , 667 , 594 , 549 , 518.7])
stoppot = np.array([0.70 , 0.65 , 0.41 , 0.28 , 0.16])
m , c = np.polyfit(freq , stoppot , 1)
bestfit = m*freq + c
plot.figure()
plot.scatter(freq , stoppot , label = 'Data')
plot.plot(freq , bestfit , label = 'Best fit line')\


plot.xlabel('frequency in (Thz)')
plot.ylabel('stopping potential (v)')
plot.title('Frequency v/s Stopping Potetial ')
plot.legend()
plot.grid(True)
plot.savefig('Freq_vs_stoppot.png' , dpi= 100 )
plot.show()
print("the slope is " , m) 
print ("the interncept is " , c )
```
